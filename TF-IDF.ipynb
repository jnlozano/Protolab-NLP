```python
import spacy
import csv
import numpy as np 
import pandas as pd
from io import StringIO 
import string
nlp = spacy.load('en_core_web_sm') # python -m spacy download en_core_web_sm
```

```python
#Methods to filter comments 
import string
from spacy.lang.en import English
punctuations = string.punctuation

# Create our list of stopwords
stop_words = spacy.lang.en.stop_words.STOP_WORDS

# Load English tokenizer, tagger, parser, NER and word vectors
parser = English()

# Creating our tokenizer function
def spacy_tokenizer(sentence):
    # Creating our token object, which is used to create documents with linguistic annotations.
    mytokens = nlp(sentence)

    # Lemmatizing each token and converting each token into lowercase
    mytokens = [ word.lemma_.lower().strip() if word.lemma_ != "-PRON-" else word.lower_ for word in mytokens ]

    # Removing stop words
    mytokens = [ word for word in mytokens if word not in stop_words and word not in punctuations ]

    # return preprocessed list of tokens
    return mytokens

# Basic function to clean the text
def clean_text(text):
    # Removing spaces and converting text into lowercase
    return text.strip().lower()
```

```python
#Load in and read data 

comments = []
idea = []
with open('idea_comment_clear_halfed.csv', encoding = 'utf8') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    for row in readCSV:
        idea.append(row[2])
        comments.append(row[3])
        #print(row)
        #print(row[0])
        #print(row[0],row[1],row[2],)
```

```python
#Seperate data by ideas. 1 idea will hold all related comments

dcomments = []
dideas = []
i = 1
while i < len(idea)-1:
    if (idea[i] != idea[i+1]):
        dideas.append(idea[i])
        dcomments.append(comments[i])
        dcomments.append('+BREAK+')
        dideas.append('+BREAK+')
    
    elif (idea[i] == idea[i+1]):
        dideas.append(idea[i])
        dcomments.append(comments[i])
        ind = i 
        while(idea[ind] == idea[ind+1]):
            dcomments.append(comments[ind+1])
            ind += 1
        dcomments.append('+BREAK+')
        dideas.append('+BREAK+')
        i = ind
    print(i)
    i += 1
```

```python
#Seperate the data into easily parsable fractions using the keyword: '+BREAK+' from last loop

i = 0
commdf = pd.DataFrame(columns=['Comments'])
while i < len(dcomments):
    text = []
    ind = i 
    while(dcomments[ind] != '+BREAK+'):
        x = '' + dcomments[ind] + '' 
        text.append(x)
        ind += 1
    commdf = commdf.append({'Comments': text}, ignore_index=True)
    i = ind
    i += 1

i = 0
ideadf = pd.DataFrame(columns=['Idea'])
while i < len(dideas):
    text = []
    ind = i 
    while(dideas[ind] != '+BREAK+'):
        text.append(dideas[i])
        ind += 1
    ideadf = ideadf.append({'Idea': text}, ignore_index=True)
    i = ind
    i += 1
    
combined = pd.concat([ideadf, commdf], axis=1)
```

```python
#Ensure string is how it is meant to look. Provide count and tokens for a comment 

string = ' '.join(combined["Comments"][1])
doctest = nlp(string)

token_list = []
for token in doctest:
    token_list.append(token.text)
print(token_list)
count = len(token_list)
print(count)
```

```python
# Find longest comment section 
i = 0 
longest = ''
index = 0 
while i < len(combined):
    string = ' '.join(combined["Comments"][i])
    if len(string) > len(longest):
        longest = string
        index = i 
    i += 1

print(len(longest))
print(longest)
print('\n')
print(combined["Idea"][index])
```

```python
#Cleaning the data first 
def listToString(s):  
    
    # initialize an empty string 
    str1 = "" 
    # traverse in the string   
    for ele in s: 
        str1 += ' '
        str1 += ele   
       
    return str1 

i = 0 
while i < len(longc):
     longc[i] = spacy_tokenizer(longc[i]) 
     longc[i] = listToString(longc[i])
     i += 1
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer 

vectorizer = CountVectorizer().fit_transform(longc)
vectors = vectorizer.toarray()

csim = cosine_similarity(vectors)
csim
```

```python
def cosine_sim_vectors(vec1,vec2):
    vec1 = vec1.reshape(1,-1)
    vec2 = vec2.reshape(1,-1)
    return cosine_similarity(vec1,vec2)[0][0]
```

```python
#find most common and least similar 
i = 0 
least = [] 
leasti = []
most = []
mosti = []
while i < len(longc) - 1:
    x = i+1 
    #y = i-1
    while x < (len(longc)):
        if (cosine_sim_vectors(vectors[i],vectors[x])) < .10:
            string = str(i) 
            string = string + ','
            string = string + str(x)
            least.append(cosine_sim_vectors(vectors[i],vectors[x]))
            leasti.append(string)
        if (cosine_sim_vectors(vectors[i],vectors[x])) > .60:
            string = str(i) 
            string = string + ','
            string = string + str(x)
            most.append(cosine_sim_vectors(vectors[i],vectors[x]))
            mosti.append(string)
        x += 1
    i += 1
```

```python
#At Sentence level in the longest comment thread
MostSim = pd.DataFrame(list(zip(most, mosti)), 
               columns =['Value', 'Index']) 
LeastSim =  pd.DataFrame(list(zip(least, leasti)), 
               columns =['Value', 'Index']) 
MostSim
```

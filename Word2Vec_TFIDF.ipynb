```python
#Load in and read data 

comments = []
idea = []
with open('idea_comment_clear_halfed.csv', encoding = 'utf8') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    for row in readCSV:
        idea.append(row[2])
        comments.append(row[3])
        #print(row)
        #print(row[0])
        #print(row[0],row[1],row[2],)
```

```python
#Seperate data by ideas. 1 idea will hold all related comments

dcomments = []
dideas = []
i = 1
while i < len(idea)-1:
    if (idea[i] != idea[i+1]):
        dideas.append(idea[i])
        dcomments.append(comments[i])
        dcomments.append('+BREAK+')
        dideas.append('+BREAK+')
    
    elif (idea[i] == idea[i+1]):
        dideas.append(idea[i])
        dcomments.append(comments[i])
        ind = i 
        while(idea[ind] == idea[ind+1]):
            dcomments.append(comments[ind+1])
            ind += 1
        dcomments.append('+BREAK+')
        dideas.append('+BREAK+')
        i = ind
    print(i)
    i += 1
```

```python
#Seperate the data into easily parsable fractions using the keyword: '+BREAK+' from last loop

i = 0
commdf = pd.DataFrame(columns=['Comments'])
while i < len(dcomments):
    text = []
    ind = i 
    while(dcomments[ind] != '+BREAK+'):
        x = '' + dcomments[ind] + '' 
        text.append(x)
        ind += 1
    commdf = commdf.append({'Comments': text}, ignore_index=True)
    i = ind
    i += 1

i = 0
ideadf = pd.DataFrame(columns=['Idea'])
while i < len(dideas):
    text = []
    ind = i 
    while(dideas[ind] != '+BREAK+'):
        text.append(dideas[i])
        ind += 1
    ideadf = ideadf.append({'Idea': text}, ignore_index=True)
    i = ind
    i += 1
    
combined = pd.concat([ideadf, commdf], axis=1)
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stop_words = set(stopwords.words('english'))
from nltk.stem import WordNetLemmatizer 
lemmatizer = WordNetLemmatizer() 
tokenizer = nltk.RegexpTokenizer(r"\w+") #removes punctuation 

def nltk_tokenizer(sent):
    
    sent = sent.lower() 
    word_tokens = tokenizer.tokenize(sent)
  
    filtered_sentence = [w for w in word_tokens if not w in stop_words] 
  
    end = []
    
    for x in filtered_sentence:
        n = lemmatizer.lemmatize(x)
        end.append(n)
    #return(word_tokens) 
    return(end) 
```

```python
#This creates a bag of words for all of the comments in a clean manner. no punctuation, lemmatized, no stop words, etc 
def listToString(s):  
    
    # initialize an empty string 
    str1 = "" 
    # traverse in the string   
    for ele in s: 
        str1 += ele   
        str1 += ' '
       
    return str1 

cleanc = longc
i = 0 
while i < len(cleanc):
    cleanc[i] = listToString(cleanc[i])
    i+=1
i = 0 


char2remove = "Ã‚"
i=0
while i < len(cleanc):
    for character in char2remove:
        cleanc[i] = cleanc[i].replace(character, "")
    i+= 1
```

```python
mydict = corpora.Dictionary([simple_preprocess(line) for line in cleanc])
corpus = [mydict.doc2bow(simple_preprocess(line)) for line in cleanc]

#for doc in corpus:
#    print([[mydict[id], freq] for id, freq in doc])

tfidf = models.TfidfModel(corpus, smartirs='ntc')
for doc in tfidf[corpus]:
    print([[mydict[id], np.around(freq, decimals=2)] for id, freq in doc])
```

```python
#pre-processing further based on gensim tutorial 

from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
for idx in range(len(cleanc)):
    cleanc[idx] = tokenizer.tokenize(cleanc[idx])
    
# Remove numbers, but not words that contain numbers.
docs = [[token for token in doc if not token.isnumeric()] for doc in cleanc]

# Remove words that are only one character.
docs = [[token for token in doc if len(token) > 1] for doc in cleanc] #Maybe not filter 'I'

#create dictionary that shows words that are filtered out
```

```python
from gensim.corpora import Dictionary

# Create a dictionary representation of the documents.
dictionary = Dictionary(docs)

# Filter out words that occur less than 10 documents, or more than 50% of the documents.
dictionary.filter_extremes(no_below=10, no_above=0.5)
```

```python
corpus = [dictionary.doc2bow(doc) for doc in docs]
#corpus
```

```python
print('Number of unique tokens: %d' % len(dictionary))
print('Number of documents: %d' % len(corpus))
```

```python
from gensim.models import LdaModel

# Set training parameters.
num_topics = 10
chunksize = 2000
passes = 20
iterations = 400
eval_every = None  # Don't evaluate model perplexity, takes too much time.

# Make a index to word dictionary.
temp = dictionary[0]  # This is only to "load" the dictionary.
id2word = dictionary.id2token

model = LdaModel(
    corpus=corpus,
    id2word=id2word,
    chunksize=chunksize,
    alpha='auto',
    eta='auto',
    iterations=iterations,
    num_topics=num_topics,
    passes=passes,
    eval_every=eval_every
)
```

```python
top_topics = model.top_topics(corpus) #, num_words=20)

# Average topic coherence is the sum of topic coherences of all topics, divided by the number of topics.
avg_topic_coherence = sum([t[1] for t in top_topics]) / num_topics
print('Average topic coherence: %.4f.' % avg_topic_coherence)

from pprint import pprint
pprint(top_topics)


#increase keywords, decrease topic size 
#tweak the filtering. too filtered right now 
#same analysis, but at idea level. for each topic, what documents/lines are contributing to this array 
#tag data in csv. relation to topics here? 


#Main focus:
#analysis across posts, grouped by similiarity or by topic. LDA/cosine similairty
#topics on topic or off topic. topic coherence. on topic with initial idea 
```

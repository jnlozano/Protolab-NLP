```python
import nltk
import csv
import numpy as np 
import pandas as pd
from nltk.corpus import stopwords
from gensim import models
import numpy as np
import gensim
#nltk.download()
#this will open a GUI from which you can download the corpus
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stop_words = set(stopwords.words('english'))
from nltk.stem import WordNetLemmatizer 
lemmatizer = WordNetLemmatizer() 
tokenizer = nltk.RegexpTokenizer(r"\w+") #removes punctuation 

def nltk_tokenizer(sent):
    
    sent = sent.lower() 
    word_tokens = tokenizer.tokenize(sent)
    
    excluded = [w for w in word_tokens if w in stop_words]

    filtered_sentence = [w for w in word_tokens if not w in stop_words] 
  
    end = []
    
    for x in filtered_sentence:
        n = lemmatizer.lemmatize(x)
        end.append(n)
    #return(word_tokens) 
    return(end,excluded) 
```

```python
#This creates a bag of words for all of the comments in a clean manner. no punctuation, lemmatized, no stop words, etc 
def listToString(s):  
    
    # initialize an empty string 
    str1 = "" 
    # traverse in the string   
    for ele in s: 
        str1 += ele   
        str1 += ' '
       
    return str1 

cleant = text
i = 0 
while i < len(cleant):
    cleant[i] = listToString(cleant[i])
    i+=1
i = 0 
```

```python
#pre-processing further based on gensim tutorial -- best used with all comments

from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')  # just removes symbols 
for idx in range(len(cleant)):
    cleant[idx] = tokenizer.tokenize(cleant[idx])
    
# Remove numbers, but not words that contain numbers. 
#docs = [[token for token in doc if not token.isnumeric()] for doc in cleani]

# Remove words that are only one character.
excluded2 = [[token for token in doc if len(token) <= 1 and token == 'i'] for doc in cleant]
docs = [[token for token in doc if len(token) > 1 and token != 'i'] for doc in cleant] #Maybe not filter 'I'

#create dictionary that shows words that are filtered out
```

```python
from gensim.corpora import Dictionary

# Create a dictionary representation of the documents.
dictionary = Dictionary(docs)

# Filter out words that occur less than 50 documents, or more than 70% of the documents. 7508 'docs' 
#dictionary.filter_extremes(no_below=50  , no_above=0.7) 
```

```python
corpus = [dictionary.doc2bow(doc) for doc in docs]
#corpus
```

```python
print('Number of unique tokens: %d' % len(dictionary))
print('Number of documents: %d' % len(corpus))
```

```python
from gensim.models import LdaModel
from gensim.models.coherencemodel import CoherenceModel

# Set training parameters. Unsure on parameters 
num_topics = 10
chunksize = 2000
passes = 20
iterations = 400
eval_every = None  # Don't evaluate model perplexity, takes too much time.

# Make a index to word dictionary.
temp = dictionary[0]  # This is only to "load" the dictionary.
id2word = dictionary.id2token

model = LdaModel(
    corpus=corpus,
    id2word=id2word,
    chunksize=chunksize,
    alpha='auto',
    eta='auto',
    iterations=iterations,
    num_topics=num_topics,
    passes=passes,
    eval_every=eval_every
)

cm = CoherenceModel(model=model, corpus=corpus, coherence='u_mass')
```

```python
top_topics = model.top_topics(corpus) #, num_words=20)

# Average topic coherence is the sum of topic coherences of all topics, divided by the number of topics.
avg_topic_coherence = sum([t[1] for t in top_topics]) / num_topics
print('Average topic coherence: %.4f.' % avg_topic_coherence)

from pprint import pprint
pprint(top_topics)

#LSI, NMF
#filter all 'problems' from ideas and run topic modeling
#classifiers for deign-orientated language. problem vs solution. 
#heuristics: it should, it would, it could.. etc for solutions. 'how might we' for problem. What language would = problem, solution
#Topic Coherence: cosine similiarty still important for canvas data

#look into visualization (email) 
#write a google doc with sections on current analysis on canvas. what do we want to do on canvas data vs twitter data
#potential analysis 
#different resources, different topic models
#language that indicates social roles (outsider vs member) 
```
